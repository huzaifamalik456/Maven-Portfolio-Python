```python
fig, ax = plt.subplots(figsize = (10,6))

ax.plot(
    utah_saledate_data.index,
    utah_saledate_data["avg_diff_to_mmr"],
    c="#cc5500",
    label = "Avg Diff to MMR"
)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
plt.xticks(rotation=30)

ax.set_xticks(utah_saledate_data.index)
ax.set_xlabel("Sale Date")
ax.set_ylabel("Avg Diff to MMR")

ax2 = ax.twinx()

#ax2.set_ylim(bottom=0)

ax2.bar(
    utah_saledate_data.index,
    utah_saledate_data["total_sold"],
    label = "Total Trucks Sold",
    color = "green", width = 5, alpha = 0.6
)

ax2.set_ylabel("Total Trucks Sold")

fig.legend(bbox_to_anchor = (.9, .88))

plt.title("Utah Truck Sales: Avg Diff to MMR vs Total Sold", fontweight = "bold", fontfamily = "serif")
plt.tight_layout();
```
