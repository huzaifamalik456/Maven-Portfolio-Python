```python
brazil = []
colombia =[]
ethiopia = []
others = []

for i in range(0,prices_paid_to_growers.shape[0],1):
    country = prices_paid_to_growers.iloc[i,0]
    
    for j in range(1,prices_paid_to_growers.shape[1],1):
        
        if country == "Brazil":
            brazil.append(float(prices_paid_to_growers.iloc[i,j]))
            
        elif country == "Colombia":
            colombia.append(float(prices_paid_to_growers.iloc[i,j]))
            
        elif country == "Ethiopia":
            ethiopia.append(float(prices_paid_to_growers.iloc[i,j]))
            
        else:
            others.append(float(prices_paid_to_growers.iloc[i,j]))
```
